### 1. Detecci√≥n de contornos

Detecte bordes en una imagen que tenga distitas formas

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
# Cargar la imagen
image = cv2.imread("./img/formas.png")

# Convertimos a escala de grises
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Reduce el ruido
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detectamos bordes
edges = cv2.Canny(blurred, 50, 150)

# Encontramos contornos
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibujamos los contornos en la imagen original
output_image = image.copy()
for contour in contours:
    epsilon = 0.02 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)
    cv2.drawContours(output_image, [approx], -1, (0, 255, 0), 3)

# Mostramos la imagen
plt.figure(figsize=(8, 8))
plt.imshow(output_image)
plt.axis("off")
plt.show()
```

```python
#EJEMPLO
# Contruyendo una imagen
x, y = np.ogrid[-np.pi:np.pi:100j, -np.pi:np.pi:100j]
r = np.sin(np.exp((np.sin(x)**3 + np.cos(y)**2)))*255
r = np.uint8(r)

out = np.zeros((100,100,3), np.uint8)
out[...,0]=r
out[...,1]=r
out[...,2]=r

# Detectamos bordes
edges = cv2.Canny(out, 120, 130)

# Encontramos contornos
contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Dibujamos los contornos en la imagen original
output_image = out.copy()
for contour in contours:
    epsilon = 0.02 * cv2.arcLength(contour, True)
    approx = cv2.approxPolyDP(contour, epsilon, True)
    cv2.drawContours(output_image, [approx], -1, (0, 255, 0), 1)

# Mostramos la imagen
plt.figure(figsize=(5, 5))
plt.imshow(output_image)
plt.axis("off")
plt.show()
```
